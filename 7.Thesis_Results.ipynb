# ROC_AUC + SHAP Plots

```python
def fun_fig_com(path1, path2, i):
    img1 = mpimg.imread(path1)
    img2 = mpimg.imread(path2)

    fig, axes = plt.subplots(nrows = 2, ncols = 1 , figsize=(20, 10))

    axes[0].imshow(img1)
    axes[0].axis('off')
    # axes[0].set_title(f'ROC_AUC df{i}', fontweight='bold')

    axes[1].imshow(img2)
    axes[1].axis('off')
    # axes[1].set_title(f'SHAP df{i}', fontweight='bold')

    plt.tight_layout()
    plt.show()
```

```python
fun_fig_com(path_roc0, path_qda0, 0)
fun_fig_com(path_roc1, path_qda1, 1)
fun_fig_com(path_roc2, path_qda2, 2)
fun_fig_com(path_roc3, path_qda3, 3)
fun_fig_com(path_roc4, path_qda4, 4)
fun_fig_com(path_roc5, path_qda5, 5)
```
