# Input Df

```python
dfis = [df0, df1, df2, df3, df4, df5, df]
iss = [0, 1, 2, 3, 4, 5, '']
```

# Interactions

```python
# create Interactions for all dfis
cols1 = [col for col in df.columns if 'Intervention_Type_List' in col]
cols2 = [col for col in df.columns if 'Funder_Type_List' in col]

for i, dfi in enumerate(dfis):
    for col1, col2 in product(cols1, cols2):
        inter_name = f"{col1}_x_{col2}"
        dfi[inter_name] = dfi[col1] * dfi[col2]

    pivot = dfi.pivot_table(index = "Study_Status_Bin",
                    values = [col for col in dfi.columns if "_x_" in col],
                    aggfunc = "sum")
    
    print(f"\n--- Pivot table for df{i} ---")
    display(pivot)
```

```python
for i, dfi in zip(iss, dfis):
    y = (dfi["Study_Status_Bin"] == "COMPLETED").astype(int)   
    inter_cols = [col for col in dfi.columns if '_x_Funder_Type_List_' in col]
    # valid_cols = [col for col in inter_cols if any(y[dfi[col] == 1] != 0) or any(y[dfi[col] == 0] != 0)]
    min_count = 10  # minimum number of non-zero samples required
    valid_cols = [col for col in inter_cols if dfi[col].sum() >= min_count]
    invalid_cols = [col for col in inter_cols if dfi[col].sum() < min_count] 

    print(f'------------------------------- df{i} ---------------------------------')
    interaction_cols = [f"{c1}_x_{c2}" for c1, c2 in product(cols1, cols2)]

    y = dfi["Study_Status_Bin"]
    X_inter = dfi[interaction_cols][valid_cols].copy()
    X_inter = sm.add_constant(X_inter, has_constant="add")

    model = sm.Logit(y, X_inter)
    result = model.fit(method="lbfgs", maxiter=1000)
    print(result.summary())

    # Drop invalid columns --> Small data sample --> drop to reduce dimension too.
    dfi = dfi.drop(columns = invalid_cols, axis = 1)

```
